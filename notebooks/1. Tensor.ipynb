## CUDA Speed

```python
from datetime import datetime
```

```python
A = torch.rand(10000, 10000)
B = torch.rand(10000, 10000)
```

```python
A.device
```

```python
B.device
```

```python
start_time = datetime.now()

A @ B

end_time = datetime.now()

time_difference = (end_time - start_time).total_seconds() * 10**3
print("Execution time of program is: ", time_difference, "ms")
```

```python
A = A.to('cuda')
B = B.to('cuda')
```

```python
start_time = datetime.now()

A @ B

end_time = datetime.now()

time_difference = (end_time - start_time).total_seconds() * 10**3
print("Execution time of program is: ", time_difference, "ms")
```

# Gradient

## Example 1

```python
a = torch.tensor([0.2])
b = torch.tensor([0.4], requires_grad=True)
```

```python
a.requires_grad
```

```python
b.requires_grad
```

```python
c = a @ b

c
```

```python
c.requires_grad
```

```python
c.backward()
```

```python
a.grad
```

```python
b.grad
```

```python
a
```

```python
b
```
